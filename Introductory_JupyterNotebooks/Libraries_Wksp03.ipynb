```python
t = np.arange(0., 5., 0.2)
print(t)
```

```python
# Let's build up on some numerical practice options

fig = plt.figure(figsize=(9, 4))

t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t,    'r--', 
         t, t**2, 'bs', 
         t, t**3, 'g^')

plt.xlabel("t")
plt.ylabel("functions of t")

plt.show()
plt.clf()
del t
```

```python
data1 = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data1['b'] = data1['a'] + 10 * np.random.randn(50)
data1['d'] = np.abs(data1['d']) * 100



fig = plt.figure(figsize=(8,6))

plt.scatter(data=data1, 
            x='a',y='b',
            c='c',
            s='d')

plt.xlabel('entry a')
plt.ylabel('entry b')

plt.show()
plt.clf()
del data1
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

# Plot 1- Both graphs together
fig = plt.figure(figsize=(9,3))

plt.plot(t1, f(t1), 'bo', 
         t2, f(t2), 'k')

plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
plt.clf()
```

```python
# Plot 2- Plots separated
fig = plt.figure(figsize=(9,6))

plt.subplot(211) # if I use XYZ, X = number of rows, Y = number of columns, Z =subplot being used
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
plt.clf()


del t1, t2
```

Using matplotlib default with pandas, understanding the arguments and keywords are essential (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html):
- DataFrame: The object for which the method is called.
- x: label or position, default None
- y: label, position or list of label, positions, default None
  - Allows plotting of one column versus another. Only used if data is a DataFrame.

- kind: str; The kind of plot to produce:


| _kind_ | _description_ |
|:--:|:--:|
|‘line’ |line plot (default) |
|‘bar’  |vertical bar plot |
|‘barh’ |horizontal bar plot |
|‘hist’ |histogram |
|‘kde’ |Kernel Density Estimation plot |
|‘box’ |boxplot |
|‘density’ |same as ‘kde’ |
|‘area’ |area plot |
|‘pie’ |pie plot |
|‘scatter’ |scatter plot |
|‘hexbin’ |hexbin plot |


- ax: matplotlib axes object, default None; An axes of the current figure.
- subplots: bool, default False; Make separate subplots for each column.
- sharex: bool, default True if ax is None else False
- sharey: bool, default False
- figsize: a tuple (width, height) in inches; Size of a figure object.

- use_index; bool, default True; Use index as ticks for x axis.
- title: str or list; Title to use for the plot. If a string is passed, print the string at the top of the figure. If a list is passed and subplots is True, print each item in the list above the corresponding subplot.
- grid: bool, default None (matlab style default); Axis grid lines.
- legend: bool or {‘reverse’}; Place legend on axis subplots.
- style: list or dict; The matplotlib line style per column.
- xticks: sequence; Values to use for the xticks.
- yticks: sequence; Values to use for the yticks.
- xlim: 2-tuple/list; Set the x limits of the current axes.
- ylim: 2-tuple/list; Set the y limits of the current axes.
- xlabel: label, optional; Name to use for the xlabel on x-axis. Default uses index name as xlabel, or the x-column name for planar plots.
- ylabel: label, optional; Name to use for the ylabel on y-axis. Default will show no ylabel, or the y-column name for planar plots.
- rot: int, default None; Rotation for ticks (xticks for vertical, yticks for horizontal plots).
- fontsize: int, default None; Font size for xticks and yticks.
- colormap: str or matplotlib colormap object, default None; Colormap to select colors from. If string, load colormap with that name from matplotlib.
- colorbar: bool, optional; If True, plot colorbar (only relevant for ‘scatter’ and ‘hexbin’ plots).

```python
# Using the customization above, let's re-do that "apps", "accepts", and "enroll" figure again

fig, axes = plt.subplots(nrows=3, ncols=2,
                         figsize=(10,6), sharex=True)

df.apps.plot(color="green", ax=axes[0,0])

df.accept.plot(color="red", ax=axes[0,1])

df.enroll.plot(color="black", ax=axes[2,1])

axes[0,0].set_ylabel("Applications")
axes[0,1].set_ylabel("Acceptances")
axes[2,1].set_ylabel("Enrollments")

plt.tight_layout()
plt.show()
plt.clf()
```

Let's control the customizability a bit more directly using MatPlotLib

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 9), sharex=True)

df.apps.plot.bar(color="blue", ax=axes[0])
df.accept.plot.bar(color="red", ax=axes[1])
df.enroll.plot.bar(color="green", ax=axes[2])

plt.show()
plt.clf()
```

```python
fig= plt.figure( figsize=(10, 3) )

plt.plot(df["enroll"]/df["apps"],
         color="black", 
        alpha=0.50, 
        linestyle="--")
plt.xlim(0, len(df))

plt.show()
plt.clf()
```

## Seaborn
----------------------------------

Matplotlib has a lot of greater functionality for higher degrees of customization, but it can get a bit overwhelming for newer programmers. A great package that works nicely for newer users is the Seaborn package, developed specifically for statistical data visualization.


```python
import seaborn as sns
```

Seaborn organizes its modules as either:

| __relplot__ | __displot__ | __catplot__ |
|:-------------:|:-------------:|:-------------:|
| _relational_ | _distributions_ | _categorical_ |
| scatterplot | histplot | stripplot |
| lineplot | kdeplot | swarmplot |
| | ecdfplot | boxplot |
| | rugplot | violinplot |
| | | pointplot |
| | | barplot |


A common use of Seaborn is to make __distribution__ plots. Let's look at the distribution of graduation rates


```python
sns.displot(data=df, x="grad_rate")
```

```python
# We can use the default histogram and change the bins

sns.displot(data=df, x="grad_rate", bins=50)
```

```python
# We can fine tune our distributions based on categorical data in the data, like whether it's a public or private college

sns.displot(data=df,
            x="grad_rate",
            hue="private",
            bins=50)
```

```python
# This can get a bit messy. One fix is to stack the bars on top

sns.displot(data=df, x="grad_rate", hue="private", multiple="stack", bins=50)
```

```python
# Another is to use a kernal density, which looks like a single curve that appears to outline the histograms

sns.displot(data=df, x="grad_rate", hue="private", kind="kde")
```

Earlier, we used the pandas scatter_matrix to hunt down relationships.

We can do that again with seaborn using __pairplot__

```python
sns.pairplot(data=df[["room_Board","personal","phd","s_f_ratio"]])
plt.show()
plt.clf()
```

```python
sns.pairplot(data=df[["private","room_Board","personal","phd","s_f_ratio"]], hue="private")
plt.show()
plt.clf()
```

We can isolate two variables with seaborn in a __jointplot__

```python
sns.jointplot(data=df,
             x="grad_rate", 
             y="s_f_ratio")
plt.show()
plt.clf()
```

```python
sns.jointplot(data=df,
             x="grad_rate", 
             y="s_f_ratio",
             hue="private")
plt.show()
plt.clf()
```
